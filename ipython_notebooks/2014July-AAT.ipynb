## Code for investigating possible hosts

```python
"""
Yao made the following cuts:
- M_K between -23 and -24.5
- M_r < -18   (this removes strange things and selects for SDSS)
- distance between 20 and 45 Mpc
- galactic latitude > 25 or < -25
- outside Rvir of any halo whose mass > 5e12 (according to the 2MASS catalog)
"""
from masterlist import masterlist

ml = ascii.read('newmasterlist.csv')

dist = ml['distance'][:]
dist[dist.mask] = ml['vhelio'][dist.mask]/70.4
print dist.mask.sum()
dm = 5* np.log10(dist) + 25

M_K = ml['K'] - dm
ml.add_column(MaskedColumn(name='M_K', data=M_K))
M_r = ml['r'] - dm
ml.add_column(MaskedColumn(name='M_r', data=M_r))
c_gal = ICRS(ml['RA']*u.deg, ml['Dec']*u.deg).transform_to(Galactic)

msk = np.ones(len(ml), dtype=bool)
msk = msk & (-24.5 < M_K) & (M_K < -23)
msk = msk & (20 < dist) & (dist < 42)
#only mask those that have an r - allow those *not* in NSA for now:
#msk[~M_r.mask] = msk[~M_r.mask] & (M_r[~M_r.mask] < -18)
#this disallows all not in nsa
msk = msk & (M_r < -18)
msk = msk & ((-25*u.deg > c_gal.b) | (c_gal.b > 25*u.deg))

hl0 = ml[msk]
hl = masterlist.remove_from_list(hl0, 'host', erroronnomatch=False,tol=.3*u.arcsec)
```
